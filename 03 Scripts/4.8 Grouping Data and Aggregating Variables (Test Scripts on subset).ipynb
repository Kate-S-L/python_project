# 3. Practicing grouping techniques

```python
#creating a subset of the data
df = ords_prods_merge[:1000000]
```

```python
#check shape of df
df.shape
```

```python
#testing out the pandas groupby() function
df.groupby('product_name')
```

```python
#finding the mean number of orders for each department_id (grouped by department)
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#doing the same without the aggregate function
df.groupby ('department_id')['order_number'].mean()
```

```python
#performing multiple aggregations
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

# 4. Create a loyalty flag using transform() and loc()

```python
#use transform and groupby() to find the maximum order of each user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#check result
ords_prods_merge.head(15)
```

```python
#prevent pandas from limiting the number of rows printed
pd.options.display.max_rows = None
```

```python
#checking transform script is correct
ords_prods_merge.head(100)
```

```python
#creating a loyalty flag with loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
