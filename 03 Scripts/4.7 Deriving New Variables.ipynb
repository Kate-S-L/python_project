# 3. Creating price flags

Create a price flag using if statements

```python
#create a subset of the data
df = df_whole_merged[:1000000]
```

```python
#check shape of subset
df.shape
```

```python
#check head of subset
df.head()
```

```python
#creating a new user-defined function 'price_label' to split orders into price categories

def price_label(row): 
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range Product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
#apply the function and create a new column to store 'price_label'
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
#check the frequency of the price_range column
df['price_range'].value_counts()
```

```python
#check the head of the df with the new column
df.head()
```

```python
#Check what the most expensive product in the subset is
df['prices'].max()
```

Create a price flag using loc() function on the whole dataset

```python
#create price flag with loc in new column 'price_range_loc'
df_whole_merged.loc[df_whole_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_whole_merged.loc[(df_whole_merged['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_whole_merged.loc[df_whole_merged['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
#check frequency of new column 'price_range_loc'
df_whole_merged['price_range_loc'].value_counts(dropna = False)
```

# 4. Creating a flag for busy and less busy days

```python
#checking which days are busiest
df_whole_merged['orders_days_of_the_week'].value_counts(dropna = False)
```

```python
#creating flag using for-loop
result = []

for value in df_whole_merged['orders_days_of_the_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy day')
    else:
        result.append('Regularly busy')
```

```python
#print result
result
```

```python
#creating a new column for the result list
df_whole_merged['Busiest day'] = result
```

```python
#checking frequency of 'busiest day' column
df_whole_merged['Busiest day'].value_counts(dropna = False)
```

```python
#Create a second column for the two busiest and two least busy days
#create the for-loop
result_2 = []

for value in df_whole_merged['orders_days_of_the_week']:
    if value == 0 or value == 1:
        result_2.append('Two busiest days')
    elif value == 4 or value == 3:
        result_2.append('Two least busy days')
    else:
        result_2.append('Regularly busy')
```

```python
df_whole_merged['Two busiest days'] = result_2
```

```python
#print list result_2
result_2
```

```python
#3 Checking the frequency
df_whole_merged['Two busiest days'].value_counts(dropna = False)
```

```python
df_whole_merged.head()
```

From the frequency count we notice that there are fewer less busy days, than there are busy days. 

# 5. Creating a flag for busiest hours of the day

```python
#Checking frquency of order time
df_whole_merged['time_of_order_24hr_time'].value_counts(dropna = False)
```

```python
#creating the for-loop
result_3 = []

for value in df_whole_merged['time_of_order_24hr_time']:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
        result_3.append('Most orders')
    elif value == 17 or value == 8 or value == 18 or value == 19 or value == 20 or value == 7 or value == 21 or value == 22:
        result_3.append('Average orders')
    else:
        result_3.append('Fewest orders')
```

```python
#print result_3
result_3
```

```python
#creating a new column
df_whole_merged['busiest_period_day'] = result_3
```

```python
#5 Checking the frequency of the new column
df_whole_merged['busiest_period_day'].value_counts(dropna = False)
```
