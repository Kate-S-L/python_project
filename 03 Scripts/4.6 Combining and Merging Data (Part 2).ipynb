```python
#Importing merged data set from previous notebook: orders_products_combined.pkl
merged_1 = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
#importing products_checked.csv
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'))
```

# 3. Combining Two Dataframes

```python
#combine 2 dfs using merge
merged_2 = merged_1.merge(df_prods, on = 'product_id', indicator = True)
```

```python
#call merged_2
merged_2
```

```python
#checking whether dfs have full match
merged_2['_merge'].value_counts()
```

```python
#drop indicator column
merged_2.drop(columns = ['_merge'], inplace = True)
```

# 4. Exporting the dfs

```python
#Exporting the data as .pkl
merged_2.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'final_combined.pkl'))
```

```python
#exporting the data as .csv
merged_2.to_csv(os.path.join(path, '02 Data','Prepared Data', 'final_combined.csv'), index = False)
```
