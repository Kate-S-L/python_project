## 03. Updating Busiest Day Column

```python
#Print frequency of “orders_day_of_week” column

df ['orders_day_of_week'].value_counts(dropna = False)
```

**Two busiest days of week are 0 & 1 (Saturday & Sunday). Two slowest days of week are 3 & 4 (Tuesday & Wednesday)

```python
#Update 'busiest day' & 'least busy' categories to 'busiest days' & 'least busiest days' using For-loop

result = []

for value in df["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest days")
  elif value == 1:
    result.append("Busiest days") 
  elif value == 3:
    result.append("Least busiest days")
  elif value == 4:
    result.append("Least busiest days")
  else:
    result.append("Regularly busy")
```

```python
#Seeing result of loop
result
```

```python
#Create new column w/in df & set equal to result

df ['busiest_days'] = result
```

```python
#Print frequency of this new column
df ['busiest_days'].value_counts(dropna=False)
```

**Frequency of busiest_days columun does match frequency of “orders_day_of_week” column

## 04. Identifying Busiest Hours of the Day Column

```python
#Print frequency of “orders_day_of_week” column

df ['order_hour_of_day'].value_counts(dropna = False)
```

**Top five busiest hours of day are 10, 11, 13, 14 & 15. Five slowest hours of day are 1, 2, 3, 4 & 5

```python
#Create new column “busiest_period_of_day" w/ time periods labels using For-loops

result = []

for value in df["order_hour_of_day"]:
  if value <= 5:
    result.append("Fewest orders")
  elif(value >=10) and (value <= 15):
    result.append("Most orders") 
  elif ((value >= 6) and (value <=9)) or (value ==12) or (value>=16):
    result.append("Average orders")
  else:
    result.append("No Data")
```

```python
#Seeing result of loop
result
```

```python
#Create new column w/in df & set equal to result

df ['busiest_period_of_day'] = result
```

```python
#Print frequency of this new column

df ['busiest_period_of_day'].value_counts(dropna=False)
```

**Frequency of busiest_period_of_day columun does match frequency of “order_hour_of_day” column
