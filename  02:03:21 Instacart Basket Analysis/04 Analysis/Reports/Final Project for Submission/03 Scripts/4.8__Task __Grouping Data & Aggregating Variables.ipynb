## 03. Find aggregated mean of “order_number” column grouped by “department_id” for entire df

```python
#Split data into groups based on “department_id.” & Aggregate each group for mean values of “order_number” column

ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

More departments are displayed for entire df vs subset. The majority of entire df means appear LOWER vs. subset means.

## 04. Difference b/w spending habits of 3 types of customers?

```python
#Split data into groups based on “loyalty_flag” & aggregate each group for mean, max & min values of “price” column

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

On average Loyal customers purchase lower priced items while new customers purchase higher priced items (with regular customers slightly lower than new customers)

## 05. Create spending flag for each user based on average price across all orders

```python
#Create new 'mean_price_ords' column based on User ID

ords_prods_merge['mean_price_ords'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Ensure column created
ords_prods_merge.head(5)
```

```python
#Create Categories based on mean_price_ords

ords_prods_merge.loc[ords_prods_merge['mean_price_ords'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price_ords'] >= 10, 'spender_flag'] = 'High spender'
```

```python
#Ensure flag column created

ords_prods_merge.head(5)
```

```python
#Check frequency

ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

## 06. Create order frequency flag for each user based on days_since_prior_order

```python
#Create new 'order_frequency' column based on days_since_prior_order

ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Ensure order_frequency column created
ords_prods_merge.head(1)
```

```python
#Create Categories based on order_frequency

ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'frequent_customer_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <= 20), 'frequent_customer_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'frequent_customer_flag'] = 'Frequent customer'
```

```python
#Ensure frequent_customer_flag column created

ords_prods_merge.head(1)
```

```python
#Check frequency

ords_prods_merge['frequent_customer_flag'].value_counts(dropna = False)
```

```python
#Create subset of NaN to view rows

subset = ords_prods_merge[ords_prods_merge['frequent_customer_flag']== 'NaN']
```
