## 05. Pricing Line Chart

```python
#Create Sample df List holding True/False values to test np.random.rand() <=0.7

np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
#Results of Sample
dev
```

```python
#Store 70% of sample in df big, Store 30% of sample in df small

big = df[dev]
small = df[~dev]
```

```python
#Checking Size of Original df
len(df)
```

```python
#Checking Size of Big & Small df combined
len(big)+len(small)
```

```python
#Reduce samples to only columns necessary for the chart
df_2 = small[['order_hour_of_day','prices']]
```

```python
#Create line chart based on Sample df_2
line = sns.lineplot(data = df_2, x = 'order_hour_of_day',y = 'prices')
```

```python
#Export line charts outside of jupyter in form of image file
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_ords_hour_of_day_task_4_9_part2.png'))
```

## 06. Line chart exploring connections b/w Age & # Dependents

```python
#Reduce samples to only columns necessary for the chart
df_3 = small[['Customer_Age','number_dependants']]
```

```python
#Create line chart based on Sample df_3
line_2 = sns.lineplot(data = df_3, x = 'Customer_Age',y = 'number_dependants')
```

## 07. Scatterplot exploring connections b/w Age & Spending Power

```python
#Create Scatterplot

scatterplot = sns.scatterplot(x = 'Customer_Age', y = 'income', data = df)
```

Above scatterplot shows that buying power certainly increases once customers reach the age of 40. It remains fairly consistent between the ages of 40 to 80 & is fairly consistent between the ages of 16-40.

```python
#Export scatterplot outside of jupyter in form of image file
scatterplot.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scatterplot_age_income_task_4_9_part2.png'))
```
