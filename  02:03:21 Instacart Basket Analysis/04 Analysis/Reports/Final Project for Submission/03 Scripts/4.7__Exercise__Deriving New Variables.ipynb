## 04. Categorizing Products by Price Range using User-Defined Functions

```python
#Create subset with first 1 million rows

df = ords_prods_merged [:1000000]
```

```python
df.shape
```

```python
#Defining the Price Function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#Applying Price Function on df & creating new column title price_range

df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Checking Values in New Column

df ['price_range'].value_counts(dropna = False)
```

```python
#Finding most expensive product within the subset 

df['prices'].max()
```

## 05. Categorizing Products by Price Range using loc() Function on Subset

```python
df.loc[df['prices']>15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Checking Values in New Column

df['price_range_loc'].value_counts(dropna = False)
```

Same value counts using user-defined vs loc() functions

## 07. For-Loops 

```python
#For-Loops Practice
for x in range(30,45):
    print("My age is %d" %(x))
```

```python
#Print frequency of “orders_day_of_week” column

ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

**O is Saturday (busiest day of week) & 4 is Wednesday (slowest day of week)

```python
#Create a new column “busiest day" using For-loop

result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Seeing result of loop
result
```

```python
#Create new column w/in ords_prods_merge & set equal to result

ords_prods_merged['busiest_day'] = result
```

```python
#Print frequency of this new column

ords_prods_merged['busiest_day'].value_counts(dropna=False)
```
